```python
%matplotlib widget
import rebound

sim = rebound.Simulation()
sim.add("10") #Sun 0
sim.add("599") #Jupyter 1

#Juypter
sim.add("502", primary=sim.particles[1]) #Europa 2
sim.add("501", primary=sim.particles[1]) #Io 3
sim.add("504", primary=sim.particles[1]) #Callisto 4

#L4
sim.add("588") #5888 Achilles L4 5
sim.add("1437") #(1437) Diomedes L4 6
sim.add("15094") #15094 Polymele (1999 WB2) 7
sim.add("11351") #11351 Leucus (1997 TS25) 8
sim.add("21900") #21900 Orus (1999 VQ10) 9

#L5
sim.add("A906 UL") #617 Patroclus (A906 UL) 10
sim.add("884") #(884) Priamus 11
sim.add("1172") #(1172) Äneas 12
sim.add("1173") #(1173) Anchises 13


zoom=7

ob1 = rebound.OrbitPlot(sim, particles = [1, 5, 6, 7, 8, 9, 10, 11, 12, 13], xlim=(sim.particles[1].x-zoom, sim.particles[1].x+zoom), ylim=(sim.particles[1].y-zoom,sim.particles[1].y+zoom), orbit_style=None)
ob2 = rebound.OrbitPlot(sim, particles = [2,3,4], primary=1, show_primary=False, fig=ob1.fig, ax = ob1.ax)

ob1.particles.set_color(["yellow", "black","black","black","black","black", "grey", "grey", "grey", "grey"])
ob2.particles.set_color(["blue"])

w = sim.getWidget()
w
```

```python
fig = ob1.fig
for i in range(100):
    ob1.sim.integrate(sim.t+0.1)
    ob2.sim.integrate(sim.t+0.1)

    ob1.xlim = [sim.particles[0].x-zoom, sim.particles[0].x+zoom]
    ob1.ylim = [sim.particles[0].y-zoom, sim.particles[0].y+zoom]

    ob1.update()
    ob2.update()

    fig.canvas.draw()
    w
```
